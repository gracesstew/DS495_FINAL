```python
jan18['HST: Hawaii Standard Time (UTC -10:00)'] = pd.to_datetime(
    jan18['HST: Hawaii Standard Time (UTC -10:00)'],
    format='%Y-%m-%d %H:%M:%S',  # specify the exact format
    errors='coerce'               # invalid values become NaT
)
jan18.head()

jan18['month'] = jan18['HST: Hawaii Standard Time (UTC -10:00)'].dt.to_period('M')
jan18_avg = jan18.groupby('month')[['Water Temperature (degF)', 'Salinity (PSU)', 'Turbidity (NTU)', 'Chlorophyll (ug/L)']].mean().reset_index()
print(jan18_avg)

```

```python
def monthly_avg(df, date_col, value_cols):
    df[date_col] = pd.to_datetime(df[date_col], format='%Y-%m-%d %H:%M:%S', errors='coerce')
    df['month'] = df[date_col].dt.to_period('M')
    return df.groupby('month')[value_cols].mean().reset_index()
```

```python
feb18_avg = monthly_avg(
    feb18,
    date_col='HST: Hawaii Standard Time (UTC -10:00)',
    value_cols=['Water Temperature (degF)', 'Salinity (PSU)', 'Turbidity (NTU)', 'Chlorophyll (ug/L)']
)
print(feb18_avg)
```
